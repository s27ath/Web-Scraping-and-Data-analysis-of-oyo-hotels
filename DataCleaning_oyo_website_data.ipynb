Outlier detection:

```python
def outlier_detection(col,df_name):
    Q1 = np.quantile(col,0.25)
    Q2 = np.quantile(col,0.50)
    Q3 = np.quantile(col,0.75)
    # calculating iqr
    IQR = Q3-Q1
    # lower boundary
    low_boundary = Q1-(1.5*IQR)
    # upper boundary
    upper_boundary = Q3+(1.5*IQR)
    result = df_name.loc[(col<low_boundary)|(col>upper_boundary)]
    print(f"Total number of outliers is {len(result)}")
    print(len(result))
    return result
```

```python
# Finding the number of outliers in 'Original Price' column
outlier_detection(data['Original Price'],data)
```

```python
# Finding the number of outliers in 'Final Price' column
outlier_detection(data['Final Price'],data)
```

```python
# Finding the number of outliers in 'Ratings' column
outlier_detection(data['Ratings'],data)
```

```python
# Finding the number of oultiers in 'Ratings Count' column
outlier_detection(data['Ratings Count'],data)
```

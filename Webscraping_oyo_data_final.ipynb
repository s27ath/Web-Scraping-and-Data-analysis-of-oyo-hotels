Importing the required libraries:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.request import Request, urlopen
```

Extracting the details of OYO Hotel Rooms in Hyderabad:

```python
url = "https://www.oyorooms.com/hotels-in-hyderabad/"
hotel_names =[]
pricings=[]
ratings = []
reviewCounts = []
wizardmem = []
ratingssummary=[]
originalprices=[]
locations=[]

def getHotelDetails(url):
    # page = requests.get(url)
    req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
   
    # text = page.text
    soup = BeautifulSoup(webpage,"html.parser")
    hotel_list = soup.find_all('div', 'oyo-row oyo-row--no-spacing listingHotelDescription')

    #Extracting hotel names
    for hotel in hotel_list:
        contact = hotel.find('div','oyo-cell--12-col listingHotelDescription__content')
        hotelname = contact.find('a','c-nn640c u-width100')
        hotel_names.append(hotelname.text)

    # Extracting hotel room prices
    for pricingVal in hotel_list:
        pricingContainer = pricingVal.find('div','oyo-cell--12-col listingHotelDescription__priceBtn')
        pricing = pricingContainer.find('span','listingPrice__finalPrice')
        if pricing == None:
            pricings.append(np.nan)
        else:
            pricings.append(pricing.text)

    # Extracting Rating of hotel
    for rate in hotel_list:
        rating = rate.find('meta',{'itemprop':'ratingValue'})
        if rating == None:
            ratings.append(np.nan)
        else:
            ratings.append(rating.get('content'))

    # Extracting total number of Ratings
    for review in hotel_list:
        reviewCount = review.find('meta',{'itemprop':'reviewCount'})
        if reviewCount ==None:
            reviewCounts.append(np.nan)
        else:
            reviewCounts.append(reviewCount.get('content'))
    
    # Extracting Rating summary
    for rev_sum in hotel_list:
       review_summary=rev_sum.find_all('span', class_="hotelRating__ratingSummary")
       if len(review_summary)!=2:
         ratingssummary.append(np.nan)
       else:
         ratingssummary.append(review_summary[1].text)  

    # Extracting wizard members
    for wizardmember in hotel_list:
      label=wizardmember.find('div', class_="c-1vsluwb")
      if label==None:
        wizardmem.append(np.nan)
      else:
        wizardmem.append(label.text)  
    
    # Extracting Original Price
    for originalprice in hotel_list:
      price=originalprice.find('span', class_="listingPrice__slashedPrice d-body-lg")
      if price==None:
        originalprices.append(np.nan)
      else:
        originalprices.append(price.text)

    # Extracting Address
    for address in hotel_list:
      addr=address.find('span',class_="u-line--clamp-2")
      locations.append(addr.text) 

for i in range(0,7):
    print("---processing page {pg}".format(pg =i))
    url = "https://www.oyorooms.com/hotels-in-hyderabad/?page={pageNumber}".format(pageNumber = i)
    getHotelDetails(url)

data = {'Hotel Name': hotel_names,
        'List Price': pricings,
        'Ratings' : ratings,
        'Ratings value': reviewCounts,
        'Ratings Summary':ratingssummary,
        'Wizard Member':wizardmem,
        'Original Price':originalprices,
        'Location':locations}
# creating dataframe       
df = pd.DataFrame(data)
print(df)
# converting into csv file
df_hyd=df.to_csv('hyd_data.csv')
```

**Data Analysis**:

Importing required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the dataset

```python
data=pd.read_csv('/content/Oyohotelsdata.csv')
```

```python
data.info()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

Q1. Determine the average Final price for each city.

```python
# Group by city and calculate the average final price
avg_price = data.groupby('City')['Final Price'].mean()

# Print the average price for each city
print(avg_price)
```

Q2. Determine the average Original Price for each city.

```python
# Group by city and calculate the average Original price
avg_price = data.groupby('City')['Original Price'].mean()

# Print the average price for each city
print(avg_price)
```

Q3. Determine the average Ratings and Ratings count for each city.

```python
# Group by city and calculate the average Ratings
avg_ratings=data.groupby('City')['Ratings'].mean()
# print the average Ratings for each city
print(avg_ratings)

# Group by city and calculate the average Ratings Count
avg_ratings_cnt=data.groupby('City')['Ratings Count'].mean()
# print the average Ratings Count for each city
print(avg_ratings_cnt)
```

Q4. Determine the average discounts for each city.

```python
# Calculate the discount percentage and add it as a new column
data['Discount'] = ((data['Original Price'] - data['Final Price']) / data['Original Price']) * 100

# Group by city and calculate the average discount
avg_discount = data.groupby('City')['Discount'].mean()

# Print the average discount for each city
print(avg_discount)
```

Q5. Determine the count of Wizard Members and Non Wizard Members for each city.

```python
# Group by city and calculate the count of Wizard Members
wizard_cnt = data.groupby('City')['Wizard Member'].apply(lambda x: x[x == 'WIZARD MEMBER'].count())

# Print the count of Wizard Members for each city
print(wizard_cnt)

# Group by city and calculate the count of Non Wizard Members
non_wizard_cnt = data.groupby('City')['Wizard Member'].apply(lambda x: x[x == 'NON WIZARD MEMBER'].count())

# Print the count of Non Wizard Members for each city
print(non_wizard_cnt)
```

Q6. Determine the Number of hotels for each city.

```python
# Group by city and calculate the count of hotels
hotels_cnt = data.groupby('City')['Hotel Name'].count()

# Print the count of hotels for each city
print(hotels_cnt)
```

Q7. Determine the Average final price for hotels in each city if the hotel is a Wizard Member and a Non Wizard Member.

```python
data_wizard = data[data['Wizard Member'] == 'WIZARD MEMBER']

# Create a pivot table
pivot_table = pd.pivot_table(data_wizard, values='Final Price', index='City', aggfunc='mean')
print(pivot_table)
```

```python
data_Nonwizard = data[data['Wizard Member'] == 'NON WIZARD MEMBER']

# Create a pivot table
pivot_table = pd.pivot_table(data_Nonwizard, values='Final Price', index='City', aggfunc='mean')
print(pivot_table)
```

Q8. Determine the Average Ratings for hotels in each city if the hotel is a Wizard Member and a Non Wizard Member.

```python
data_wizard = data[data['Wizard Member'] == 'WIZARD MEMBER']
# Group by city and calculate the average  for wizard member
avg_ratings=data_wizard.groupby('City')['Ratings'].mean()
print(avg_ratings)
```

```python
data_Nonwizard = data[data['Wizard Member'] == 'NON WIZARD MEMBER']
# Group by city and calculate the average ratings for non wizard member
avg_ratings=data_Nonwizard.groupby('City')['Ratings'].mean()
print(avg_ratings)
```

The distribution of the Final prices using a histogram, which shows the frequency of prices within different price ranges.

Distribution of Ratings Summary using pie chart.

```python
# Compute the frequency of each rating summary in the dataset
ratingsumm_counts = data['Ratings Summary'].value_counts()
# Create a pie chart of the ratings summary
plt.pie(ratingsumm_counts, labels=ratingsumm_counts.index, autopct='%1.1f%%')
plt.title('Ratings Summary')
plt.show()
```

Interpretation from the above graph: From the above pie chart,we can see that the 35% of the  OYO hotel rooms were given the rating as very good and the very few hotel rooms were given the rating as fabulous.

Distribution of cities using pie chart.

```python
# Compute the frequency of each city in the dataset
city_counts = data['City'].value_counts()
# Create a pie chart of the cities distribution
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%')
plt.title('Cities Distribution')
plt.show()
```

Interpretation from the above graph: From the above pie chart,we can see that the major percentage of the OYO hotel rooms are available in Kolkata and very few hotel rooms are available in Mumbai.

Bivariate analysis:

Scatter plot for analysing the relation between the ratings and final price.

```python
# Extract ratings and final prices into separate arrays
ratings = data['Ratings'].values
prices = data['Final Price'].values

# Create a scatter plot of ratings vs. prices
plt.scatter(ratings, prices)

# Add labels and title
plt.xlabel('Ratings')
plt.ylabel('Final Price')
plt.title('OYO Hotels Ratings vs. Final Prices')

# Show the chart
plt.show()
```

Interpretation from above graph: from the above graph, The scatter plot shows a random distribution of the ratings and final prices for OYO hotels. However, there seems to be a slight positive correlation between the two variables, meaning that as the ratings increase, the final prices also tend to increase. This suggests that higher-rated OYO hotels may command a premium price compared to lower-rated hotels.

Bar chart of average final prices by city.

```python
# Compute the average final price by city
avg_list_price = data.groupby('City')["Final Price"].mean().reset_index()

# Create a bar chart of average final prices by city
plt.figure(figsize=(8, 6))
sns.barplot(x="City", y="Final Price", data=avg_list_price, palette="Blues_d")
plt.xticks(rotation=45, ha="right")
plt.title("Average Final Prices by City")
plt.xlabel("City")
plt.ylabel("Average Final Price")
plt.show()
```

Interpretation from the above graph: from the above graph we can see that the bar chart shows the average final prices of hotels in each city. Mumbai has the highest average final price, followed by Banglore and Kolkata. Delhi has the lowest average final price.

The distribution of Wizard Member by City using stacked bar chart.

```python
# Crosstab
ct=pd.crosstab(index=data['Wizard Member'], columns=data['City'])
# Plot the crosstab as a stacked bar chart
ct.plot(kind='bar', stacked=True)
# Set the plot title and axes labels
plt.title('Wizard Members by City')
plt.xlabel('City')
plt.ylabel('Count')
# Show the plot
plt.show()
```

Interpretation from above graph: From the above graph we can see that the proportion of non wizard member hotels are higher in Kolkata whereas the higher proportion of wizard member hotels are in Kolkata and Hyderabad.Similarly least proportion of non wizard memeber hotels is in Chennai and least proportion of wizard memeber hotels is in Mumbai.

The distribution of Ratings Summary by city using stacked bar chart.

```python
# Crosstab
ct=pd.crosstab(index=data['Ratings Summary'], columns=data['City'])
# Plot the crosstab as a stacked bar chart
ct.plot(kind='bar', stacked=True)
# Set the plot title and axes labels
plt.title('Rating Summary by City')
plt.xlabel('City')
plt.ylabel('Count')
# Show the plot
plt.show()
```

Interpretation from the above chart: from the above chart we can see that the major proportion of the Oyo hotels in Kolkata were given the rating as Very Good and the major proportion of Oyo hotels in Delhi were given the rating as Fair and in Bangalore the ratings were given as Very Good.The major proportion of Oyo hotels in Hyderabad were given the rating as Very Good and the major proportion of Oyo hotels in Mumbai were given the rating as Good.

Distribution of final prices across different cities using box plot

```python
data.boxplot(column='Final Price', by='City')
plt.title('Final Price Distribution of OYO Hotels by City')
plt.suptitle('')
plt.xlabel('City')
plt.ylabel('Final Price')
plt.show()
```

Interpretation from the above graph: From the above graph we can see that the price of oyo hotel rooms is maximum in Mumbai when compared to the other cities.The price of hotel rooms is minimum in Chennai,Hyderabad and Kolkata.

Distribution of Ratings of hotel rooms across different cities using violin plot.

```python
sns.violinplot(x='City', y='Ratings', data=data)
plt.title('Distribution of OYO Hotel Ratings by City')
plt.xlabel('City')
plt.ylabel('Ratings')
plt.show()
```

Interpretation from the above graph: from the above graph we can say that the cities Delhi and Mumbai have the least ratings for oyo hotel rooms and Hyderabad,Bangalore and Kolkata have the highest ratings for oyo hotel rooms.

Multivariate Analysis:

Determination of correlation between the features using Heat Map.

```python
# Compute the correlation matrix
corr = data.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

Interpretation from the graph: from the above graph we can say that Final Price and Original Price are highly positively correlated and Final Price and the discount are highly negatively correlated.Hence we can determine how all the features are related to each other.

```python
# Create a pairplot of the variables
sns.pairplot(data)
plt.show()
```

Interpretation from the above graph: From the graph we can see how the features are related to each other. It can also be seen that the Final Price and the original price are highly positively correlated to each other.
